## Python Basics: An Overview

Welcome to Python basics! This section will guide you through the fundamental concepts of Python programming, focusing on variables, data types, and basic operators. Understanding these core elements is crucial for building any program in Python.

Python is known for its simplicity and readability, making it an excellent language for beginners. Let's dive in!

### 1. Variables

**Theory:**

A variable is a named storage location that holds a value. Think of it as a container that can store different types of information. In Python, you don't need to declare the type of a variable explicitly; the interpreter infers it based on the assigned value.

**Key characteristics:**
*   **Dynamic Typing:** A variable can hold values of different data types during its lifetime.
*   **Assignment:** Values are assigned to variables using the `=` operator.
*   **Naming Rules:**
    *   Must start with a letter or an underscore (`_`).
    *   Cannot start with a number.
    *   Can only contain alphanumeric characters (A-Z, a-z, 0-9) and underscores.
    *   Case-sensitive (e.g., `age` is different from `Age`).
    *   Cannot be a Python keyword (e.g., `if`, `for`, `while`).

````python
# --- Practical Implementation: Variables ---

# Assigning different types of values to variables
my_integer = 10
my_float = 3.14
my_string = "Hello, Python!"
my_boolean = True

# Displaying the values of variables
print(f"Integer variable: {my_integer}")
print(f"Float variable: {my_float}")
print(f"String variable: {my_string}")
print(f"Boolean variable: {my_boolean}")

# Reassigning a variable (dynamic typing)
my_integer = "Now I am a string"
print(f"Reassigned integer variable: {my_integer}")

# Variable naming examples
_valid_name = 20
variable2 = 30
print(f"Valid name: {_valid_name}")
print(f"Another valid name: {variable2}")

# Invalid naming (would cause an error if uncommented):
# 2nd_variable = 40  # Cannot start with a number
# my-variable = 50   # Hyphens are not allowed
# class = 60         # 'class' is a keyword
````

### 2. Data Types

**Theory:**

Data types classify the type of values that a variable can hold. Python has several built-in data types, which can be broadly categorized as follows:

*   **Numeric Types:**
    *   **`int` (Integers):** Whole numbers (e.g., `5`, `-100`).
    *   **`float` (Floating-point numbers):** Numbers with a decimal point (e.g., `3.14`, `-0.5`).
    *   **`complex` (Complex numbers):** Numbers with a real and imaginary part (e.g., `3 + 4j`).
*   **Sequence Types:**
    *   **`str` (Strings):** Sequences of characters, enclosed in single or double quotes (e.g., `'hello'`, `"world"`).
    *   **`list` (Lists):** Ordered, mutable (changeable) sequences of items. Items can be of different types, enclosed in square brackets `[]` (e.g., `[1, 'apple', True]`).
    *   **`tuple` (Tuples):** Ordered, immutable (unchangeable) sequences of items, enclosed in parentheses `()` (e.g., `(1, 'banana', False)`).
*   **Mapping Type:**
    *   **`dict` (Dictionaries):** Unordered collections of key-value pairs, mutable. Keys must be unique and immutable, enclosed in curly braces `{}` (e.g., `{'name': 'Alice', 'age': 30}`).
*   **Set Types:**
    *   **`set` (Sets):** Unordered collections of unique, immutable items, mutable. Enclosed in curly braces `{}` (e.g., `{1, 2, 3}`).
    *   **`frozenset`:** An immutable version of a set.
*   **Boolean Type:**
    *   **`bool` (Booleans):** Represents truth values: `True` or `False`.

````python
# --- Practical Implementation: Data Types ---

# Numeric Types
num_int = 100
num_float = 10.5
num_complex = 3 + 2j
print(f"Integer: {num_int}, Type: {type(num_int)}")
print(f"Float: {num_float}, Type: {type(num_float)}")
print(f"Complex: {num_complex}, Type: {type(num_complex)}")

# String Type
str_example = "Python is fun!"
print(f"String: {str_example}, Type: {type(str_example)}")

# List Type (Mutable Sequence)
list_example = [1, "apple", 3.14, True]
print(f"List: {list_example}, Type: {type(list_example)}")
list_example.append(5)
print(f"List after append: {list_example}")

# Tuple Type (Immutable Sequence)
tuple_example = (10, "banana", False)
print(f"Tuple: {tuple_example}, Type: {type(tuple_example)}")
# tuple_example.append(20) # This would cause an AttributeError as tuples are immutable

# Dictionary Type (Mapping)
dict_example = {"name": "Bob", "age": 25, "city": "New York"}
print(f"Dictionary: {dict_example}, Type: {type(dict_example)}")
dict_example["age"] = 26 # Modifying a value
print(f"Dictionary after modification: {dict_example}")

# Set Type (Unordered collection of unique items)
set_example = {1, 2, 3, 2, 4}
print(f"Set: {set_example}, Type: {type(set_example)}") # Note: duplicates are removed
set_example.add(5)
print(f"Set after add: {set_example}")

# Boolean Type
bool_true = True
bool_false = False
print(f"Boolean True: {bool_true}, Type: {type(bool_true)}")
print(f"Boolean False: {bool_false}, Type: {type(bool_false)}")
````

### Basic Operators

**Theory:**

Operators are special symbols that perform operations on values and variables. Python supports various types of operators.

*   **Arithmetic Operators:** Used for mathematical calculations.
    *   `+` (Addition)
    *   `-` (Subtraction)
    *   `*` (Multiplication)
    *   `/` (Division)
    *   `%` (Modulo - returns the remainder of division)
    *   `**` (Exponentiation - power of)
    *   `//` (Floor Division - returns the integer part of the quotient)
*   **Comparison (Relational) Operators:** Used to compare two values and return a Boolean (`True` or `False`).
    *   `==` (Equal to)
    *   `!=` (Not equal to)
    *   `>` (Greater than)
    *   `<` (Less than)
    *   `>=` (Greater than or equal to)
    *   `<=` (Less than or equal to)
*   **Logical Operators:** Used to combine conditional statements.
    *   `and` (Returns `True` if both statements are true)
    *   `or` (Returns `True` if one of the statements is true)
    *   `not` (Reverses the result, returns `False` if the result is true)
*   **Assignment Operators:** Used to assign values to variables.
    *   `=` (Assigns value from right operand to left operand)
    *   `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=` (Shorthand for arithmetic operations followed by assignment)

````python
# --- Practical Implementation: Basic Operators ---

a = 15
b = 4

# Arithmetic Operators
print("\n--- Arithmetic Operators ---")
print(f"Addition (a + b): {a + b}")
print(f"Subtraction (a - b): {a - b}")
print(f"Multiplication (a * b): {a * b}")
print(f"Division (a / b): {a / b}") # Returns a float
print(f"Modulo (a % b): {a % b}") # Remainder
print(f"Exponentiation (a ** b): {a ** b}") # 15 to the power of 4
print(f"Floor Division (a // b): {a // b}") # Integer division

# Comparison Operators
print("\n--- Comparison Operators ---")
print(f"a == b: {a == b}")
print(f"a != b: {a != b}")
print(f"a > b: {a > b}")
print(f"a < b: {a < b}")
print(f"a >= b: {a >= b}")
print(f"a <= b: {a <= b}")

# Logical Operators
x = True
y = False
print("\n--- Logical Operators ---")
print(f"x and y: {x and y}")
print(f"x or y: {x or y}")
print(f"not x: {not x}")

# Assignment Operators
print("\n--- Assignment Operators ---")
c = 10
print(f"Initial c: {c}")
c += 5 # Equivalent to c = c + 5
print(f"c after c += 5: {c}")
c -= 3 # Equivalent to c = c - 3
print(f"c after c -= 3: {c}")
c *= 2 # Equivalent to c = c * 2
print(f"c after c *= 2: {c}")
c /= 4 # Equivalent to c = c / 4
print(f"c after c /= 4: {c}")
c %= 3 # Equivalent to c = c % 3
print(f"c after c %= 3: {c}")
````

This concludes the basic overview of variables, data types, and operators in Python. With these foundational concepts, you are ready to explore more advanced topics!

User Input

**Theory:**

Programs often need to interact with the user to get information. In Python, the built-in `input()` function is used to get input from the user via the console.

*   The `input()` function reads a line from input, converts it to a string, and returns it.
*   You can pass a prompt message as an argument to `input()`, which will be displayed to the user before they enter their input.
*   Remember that `input()` *always* returns a string, so you might need to convert it to another data type (e.g., `int()` or `float()`) if you're expecting numeric input.

````python
# --- Practical Implementation: User Input ---

# Get a string input from the user
user_name = input("Enter your name: ")
print(f"Hello, {user_name}!")

# Get numeric input and convert it to an integer
try:
    age_str = input("Enter your age: ")
    user_age = int(age_str)
    print(f"You are {user_age} years old.")
    if user_age >= 18:
        print("You are an adult.")
    else:
        print("You are a minor.")
except ValueError:
    print("Invalid age entered. Please enter a number.")

# Get numeric input and convert it to a float
try:
    height_str = input("Enter your height in meters: ")
    user_height = float(height_str)
    print(f"Your height is {user_height} meters.")
except ValueError:
    print("Invalid height entered. Please enter a number.")
````

Control Flow: Loops (for and while)

**Theory:**

Loops are used to execute a block of code repeatedly. Python provides two main types of loops:

*   **`for` loop:** Iterates over a sequence (like a list, tuple, string, or range) or other iterable objects. It executes a block of code for each item in the sequence.
*   **`while` loop:** Executes a block of code as long as a specified condition is `True`. It keeps looping until the condition becomes `False`.

````python
# --- Practical Implementation: Loops ---

# For loop with a list
fruits = ["apple", "banana", "cherry"]
print("\n--- For loop with a list ---")
for fruit in fruits:
    print(fruit)

# For loop with range()
print("\n--- For loop with range() ---")
for i in range(5): # Iterates from 0 to 4
    print(i)

# While loop
print("\n--- While loop ---")
count = 0
while count < 3:
    print(f"Count: {count}")
    count += 1

# Using break and continue
print("\n--- Break and Continue ---")
for i in range(10):
    if i == 3:
        continue # Skip to the next iteration if i is 3
    if i == 7:
        break    # Exit the loop if i is 7
    print(i)
````

### Control Flow: Conditional Statements (if-elif-else)

**Theory:**

Conditional statements allow your program to make decisions based on certain conditions. They execute a block of code only if a specified condition is true.

*   **`if` statement:** Executes a block of code if its condition is `True`.
*   **`elif` (else if) statement:** Checks another condition if the preceding `if` or `elif` conditions were `False`.
*   **`else` statement:** Executes a block of code if all preceding `if` and `elif` conditions are `False`.
*   **Nested `if` statements:** An `if` or `elif` statement can contain another `if-elif-else` structure within its block to handle more complex logic.

````python
# --- Practical Implementation: Conditional Statements ---

# Simple if-elif-else example
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")

# Nested if-else example
weather = "rainy"
temperature = 22

if weather == "rainy":
    if temperature < 10:
        print("It's a cold and rainy day. Bring a heavy coat!")
    elif temperature < 20:
        print("It's a cool and rainy day. Bring an umbrella and light jacket.")
    else:
        print("It's a warm and rainy day. Just an umbrella might do.")
else:
    print("It's not raining today.")
````

### 'Switch' Statement in Python (Using `if-elif-else` or Dictionary)

**Theory:**

Unlike many other programming languages (like C++, Java, JavaScript), Python does not have a built-in `switch` or `case` statement. However, you can achieve similar functionality using either a series of `if-elif-else` statements or by using a dictionary to map cases to functions or values.

**Usage with `if-elif-else` (most common equivalent):**
This is the straightforward way to handle multiple conditions, similar to how we used it for grading. Each `elif` checks a different 'case'.

**Usage with Dictionary Mapping (for cleaner code with many cases):**
For scenarios with many 'cases', a dictionary can be more elegant. You can map keys (your 'cases') to specific functions to execute or values to return.

````python
# --- Practical Implementation: 'Switch' Equivalent ---

# Example 1: Using if-elif-else for a 'switch' effect
def get_day_type_if_else(day):
    if day == "Monday":
        return "Start of the week"
    elif day == "Friday":
        return "End of the work week"
    elif day == "Saturday" or day == "Sunday":
        return "Weekend!"
    else:
        return "Mid-week day"

print(f"Monday is: {get_day_type_if_else('Monday')}")
print(f"Wednesday is: {get_day_type_if_else('Wednesday')}")
print(f"Saturday is: {get_day_type_if_else('Saturday')}")

# Example 2: Using a dictionary for a 'switch' effect
def get_day_type_dict(day):
    # Define a dictionary where keys are cases and values are functions or data
    day_actions = {
        "Monday": lambda: "Start of the week (dict)",
        "Friday": lambda: "End of the work week (dict)",
        "Saturday": lambda: "Weekend! (dict)",
        "Sunday": lambda: "Weekend! (dict)"
    }
    # Get the action for the day, or a default action if not found
    action = day_actions.get(day, lambda: "Mid-week day (dict)")
    return action()

print(f"Monday is: {get_day_type_dict('Monday')}")
print(f"Wednesday is: {get_day_type_dict('Wednesday')}")
print(f"Sunday is: {get_day_type_dict('Sunday')}")
````

### Text-to-Speech with `gTTS`

**Theory:**

Text-to-Speech (TTS) libraries allow your Python programs to convert written text into spoken audio. This can be useful for accessibility, interactive applications, or simply adding an audio dimension to your projects.

`gTTS` (Google Text-to-Speech) is a Python library and CLI tool to interface with Google Translate's text-to-speech API. It writes spoken text to an audio file. You will need to install it first.

**Installation:**

To use `gTTS`, you typically need to install it using `pip`:
```bash
pip install gTTS
```

**Usage:**

Once installed, you can import `gTTS`, create a `gTTS` object with the text and language, and then save the speech to an audio file (e.g., MP3). To play the audio in Colab, you can use `IPython.display.Audio`.

````python
!pip install gTTS
````

````python
# --- Practical Implementation: Text-to-Speech with gTTS ---

# Install gTTS library if not already installed
# !pip install gTTS

from gtts import gTTS
from IPython.display import Audio, display

# Text to be converted to speech
text_to_speak = "Hello, Colab user! This is a demonstration of Text-to-Speech using gTTS."

# Choose a language (e.g., 'en' for English)
language = 'en'

# Create a gTTS object
tts = gTTS(text=text_to_speak, lang=language, slow=False)

# Save the speech to an audio file
audio_file = "hello_colab.mp3"
tts.save(audio_file)
print(f"Speech saved to {audio_file}")

# Play the audio file in Colab
print("Playing audio...")
display(Audio(audio_file, autoplay=True))

# You can also make it speak directly using a different method if needed
# For example, to speak a shorter message:
text_to_speak_short = "Python is fun!"
tts_short = gTTS(text=text_to_speak_short, lang=language, slow=False)
tts_short.save("python_fun.mp3")
print(f"Speech saved to python_fun.mp3")
display(Audio("python_fun.mp3", autoplay=True))
````
